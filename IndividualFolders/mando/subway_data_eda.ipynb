```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/MTA_Subway_Entrances_and_Exits__2024_20241002.csv')
df.sample(10)
```

```python
entrnace_type_by_borough = df.groupby(['Borough', 'Entrance Type']).size().unstack()
```

```python
entrance_type_by_borough = df.groupby(['Borough', 'Entrance Type']).size().unstack().fillna(0)

entrance_type_by_borough.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title("Distribution of Entrance Types by Borough")
plt.xlabel("Borough")
plt.ylabel("Count of Entrances")
plt.legend(title="Entrance Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
