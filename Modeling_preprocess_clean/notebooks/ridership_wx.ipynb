```python
# Import libraries 
import pandas as pd 
import numpy as np 
```

```python
# load data sets
ridership = pd.read_csv('../../Modeling_preprocess_clean/cleaned_data/ridership_w_lag.csv')
weather = pd.read_csv('../../Modeling_preprocess_clean/cleaned_data/weather_data_ny_cleaned.csv')
```

```python
# display the first few rows of the data set and shape
display(ridership.head())
ridership.shape
```

```python
# display the first few rows of the data set and shape
display(weather.head())
weather.shape
```

```python
# ensure timestamp is in datetime format
ridership['timestamp'] = pd.to_datetime(ridership['timestamp'])
ridership['date'] = ridership['timestamp'].dt.date
```

```python
# Ensure DATE is in datetime format
weather['DATE'] = pd.to_datetime(weather['DATE'])
weather['date'] = weather['DATE'].dt.date
```

```python
# Merge the two data sets
merged_df = pd.merge(ridership, weather, how='left', on='date')
merged_df.drop(columns=['DATE'], inplace=True)
```

```python
# display the first few rows of the data set and shape
display(merged_df.head())
```

```python
# Verify the columns in the merged data set and shape
display(merged_df.columns)
merged_df.shape
```

```python
# Count all missing values in the dataset
merged_df.isnull().sum()
```

```python
# Export the merged data set 
merged_df.to_csv('../../Modeling_preprocess_clean/cleaned_data/ridership_weather.csv', index=False)
```
