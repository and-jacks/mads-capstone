```python
import pandas as pd
```

```python
events_df = pd.read_csv('data/events.csv')
events_df.head()
```

```python
# convert start and edn time to datemtime 
events_df['start_date_time'] = pd.to_datetime(events_df['start_date_time'], errors='coerce')
events_df['end_date_time'] = pd.to_datetime(events_df['end_date_time'], errors='coerce')

events_df.info()
display(events_df.head())
```

```python
# get unuique values of event type
events_df['event_type'].unique()
```

```python
# keep mlb for now
events_df = events_df[events_df['event_type'] == 'Sport - MLB']
display(events_df.head())
events_df.shape
```

```python
# get unique values of event locatoin
events_df['event_location'].unique()
```

```python
# city field gps 40° 45' 26.1216'' N and 73° 50' 46.4532'' W.
# yankee stadium  40° 49' 46.7724'' N and 73° 55' 34.2696'' W.

# for temp use.  must get gps cordiantes for each event if possible

city_field = (40.757256, -73.846237)
yankee_stadium = (40.829659, -73.926186)

# add gps column
events_df['gps'] = events_df['event_location'].apply(lambda x: city_field if x == 'Citi Field' else yankee_stadium)
events_df.head()

```

```python
# export to cleaned csv mlb 
events_df.to_csv('data/Cleaned_data/events_mlb-cleaned.csv', index=False)
```
